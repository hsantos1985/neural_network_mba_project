##2.2-SCORE 4 E 5

```python
#CRIAÇÃO DA BASE NULO 4 E 5
base_review_nulo_4_5=base_review_nulo[(base_review_nulo['review_score']==5) |(base_review_nulo['review_score']==4)]
```

```python
#PREENCHENDO A COLUNA REVIEW MESSAGE EM BRANCO DAS LINHAS COM SCORES 4 E 5
base_review_nulo_4_5['review_comment_message'] = np.where(
    ((base_review_nulo_4_5['review_score'] == 5) | (base_review_nulo_4_5['review_score'] == 4))
    &  (base_review_nulo_4_5['review_comment_message'].isnull()==True),
    'SEM_COMENTARIO_SCORE_ENTRE_4_5',
    base_review_nulo_4_5['review_comment_message']
)
```

```python
base_review_nulo_4_5['review_comment_message'].unique()
```

##2.3-SCORE 1,2,E 3

```python
#CRIAÇÃO DA BASE SCORES 1,2,3
base_review_nulo_1_2_3=base_review_nulo[(base_review_nulo['review_score']==1) |(base_review_nulo['review_score']==2)| (base_review_nulo['review_score']==3)]
```

```python
base_review_nulo_1_2_3
```

###2.1-CRIAÇÃO DA COLUNA PRAZO DAS ENTREGAS

```python
#FAZENDO O JOIN COM OLIST ORDER DATASET ITEMS
base_review_nulo_1_2_3=(
base_review_nulo_1_2_3
.merge(olist_order_dt, how='left', on='order_id', indicator=False)
.drop_duplicates()
)
```

```python
#CRIAÇÃO DA COLUNA COMPARATIVA ENTRE A DATA DE ENTREGA E A DATA DE ENTREGA PREVISTA
base_review_nulo_1_2_3['ENTREGA_PRAZO']=np.where (base_review_nulo_1_2_3['order_estimated_delivery_date']>=base_review_nulo_1_2_3['order_delivered_customer_date'],
  'No_prazo',
  'Fora_prazo'
)
```

```python
base_review_nulo_1_2_3['ENTREGA_PRAZO'].unique()
```

```python
base_review_nulo_1_2_3[base_review_nulo_1_2_3['ENTREGA_PRAZO']=='Fora_prazo']
```

```python
#PREENCHENDO AS LIHAS EM BRANCO DA REVIEW MESSAGE PARA AS ENTREGAS FORA DO PRAZO
base_review_nulo_1_2_3['review_comment_message']=np.where(
base_review_nulo_1_2_3['ENTREGA_PRAZO']=='Fora_prazo',
'RECEBEU_ENTREGA_FORA_PRAZO',
base_review_nulo_1_2_3['review_comment_message']
)
```

```python
base_review_nulo_1_2_3['review_comment_message'].unique()
```

###2.3-CRIAÇÃO DA COLUNA DO % DO PREÇO DO FRETE PELO VALOR DA COMPRA

```python
#FAZENDO OS JOINS COM AS BASES DE ORDER ITEMS E DE PAYMENTS
base_review_nulo_1_2_3=(
base_review_nulo_1_2_3
.merge(olist_order_items, how='left', on='order_id', indicator=False)
.merge(olist_payments_dataset, how='left', on='order_id', indicator=False)
.drop_duplicates()
)
```

```python
round(base_review_nulo_1_2_3['freight_value']/base_review_nulo_1_2_3['payment_value'],2)
```

```python
#CRIANDO A COLUNA DE % DO FRETE
base_review_nulo_1_2_3['PERC_FRETE_PAGAMENTO']=round(base_review_nulo_1_2_3['freight_value']/base_review_nulo_1_2_3['payment_value'],2)
```

```python
#CRIANDO A COLUNA DE CLASSIFICAÇÃO DO % DO FRETE
# Definindo as condições
condicoes = [
    base_review_nulo_1_2_3['PERC_FRETE_PAGAMENTO'] <= 0.25,
    base_review_nulo_1_2_3['PERC_FRETE_PAGAMENTO'] <= 0.5,
    base_review_nulo_1_2_3['PERC_FRETE_PAGAMENTO'] <= 0.75
]

# Definindo os valores que serão atribuídos para cada condição
valores = [
    'menor_25%',
    'menor_50%',
    'menor_75%'
]
base_review_nulo_1_2_3['FRETE_PAGAMENTO_CATEGORIA']=np.select(condicoes,valores, default='maior_75%')
```
